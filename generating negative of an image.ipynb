```python
import cv2
import matplotlib.pyplot as plt


img = cv2.imread(r"C:\Users\nidhi\OneDrive\Pictures\Mav (Camera)\20211230_104744.jpg")

img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gray = cv2.imread(r"C:\Users\nidhi\OneDrive\Pictures\Mav (Camera)\20211230_104744.jpg", 0)


colored_negative  = abs(255-img1)
gray_negative  = abs(255-gray)

imgs = [img, gray, colored_negative, gray_negative]
title = ['coloured', 'gray', 'coloured-negative', 'gray-negative']

plt.subplot(2, 2, 1)
plt.title(title[0])
plt.imshow(imgs[0])
plt.xticks([])
plt.yticks([])

plt.subplot(2, 2, 2)
plt.title(title[1])
plt.imshow(imgs[1], cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(2, 2, 3)
plt.title(title[2])
plt.imshow(imgs[2])
plt.xticks([])
plt.yticks([])

plt.subplot(2, 2, 4)
plt.title(title[3])
plt.imshow(imgs[3], cmap='gray')
plt.xticks([])
plt.yticks([])

plt.show()
```
